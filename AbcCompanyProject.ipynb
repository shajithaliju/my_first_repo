1.	How many are there in each Team and the percentage splitting with respect to the total employees.

```python
df['Team'].value_counts()
```

```python
team_counts = df['Team'].value_counts()
total_count = team_counts.sum()

# Calculate the percentage splits
team_percentage = team_counts / total_count * 100

# Display the unique values and their percentage splits
for team, percentage in zip(team_counts.index, team_percentage):
    print(f'Team: {team}, Percentage: {percentage:.2f}%')
```

2.	Segregate the employees w.r.t different positions

```python
grouped_data = df.groupby('Position')

# Iterate over the groups and display the employees in each position
for position, employees in grouped_data:
    print(f"Position: {position}")
    print(employees)
    print()
```

3.	Find from which age group most of the employees belong to.

```python
age_counts = df['Age'].value_counts()

# Get the age group with the highest count
most_common_age = age_counts.idxmax()

# Display the age group with the highest count
print(f"The most common age group among employees is: {most_common_age}")
```

4.	Find out under which team and position, spending in terms of salary is high.

```python
grouped_data = df.groupby(['Team', 'Position'])['Salary'].sum()

# Find the team and position with the highest total salary
highest_spending = grouped_data.idxmax()
team, position = highest_spending

# Display the team and position with the highest total salary
print(f"The highest spending in terms of salary is under Team: {team}, Position: {position}")
```

5.	Find if there is any correlation between age and salary , represent it visually.

```python
age_salary_corr = df['Age'].corr(df['Salary'])

# Create a scatter plot to visualize the relationship
plt.scatter(df['Age'], df['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Correlation between Age and Salary')
plt.show()

# Display the correlation coefficient
print(f"The correlation between Age and Salary is: {age_salary_corr}")
```
